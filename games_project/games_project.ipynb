## Предобработка данных.

### Приведем названия столбцов в соответствие с хорошим стилем.

```python
data.columns = data.columns.str.lower() # Приведем названия столбцов к нижнему регистру.
```

### Пропуски значений.

Посмотрим сколько в таблице пропущенных значений.

```python
data.isna().agg(['mean', 'sum']).T.sort_values(by='mean', ascending=False)
```

Пропуски в name и genre можно удалить, так как их ничтожно мало.
Пропуски значений в year_of_release занимают долю примерно в 1,6%. Мы не можем заменить год релиза на какое-либо логичное значение, поэтому посчитаем нужным удалить эти строки.
В столбцах critic_score, rating и user_score наблюдается достаточно большая доля пропусков, примерно: 51%, 40% и 40% соответственно. Удаление строк в в этих столбцах может привести к потере большого количества информации. Тем более эти параметры нужны нам для дальнейшего исследования. 
Посмотрим сколько строк занимают данные с пропусками во всех этих трех колонках.

```python
# Посмотрим срез с условием :пропуска одновременно в трех проблемных колонках.
data.query("critic_score.isna() & user_score.isna() &  rating.isna()")
```

В итоге мы имеем 6621 строку с пропусками в столбцах critic_score, rating и user_score. Можно сделать вывод, что появление пропусков в этих трех колонках взаимосвязано. Поэтому наличие этих пропусков говорит о том, что они носят неслучайный характер, и скорее всего данные для этих колонок отсутствуют, возможно игры новые, оценки еще не успели проставить. Пропусков в колонке critic_score изначально в количестве 8578, вероятно какие -то пропуска не зависели от их наличия в rating и user_score. И здесь может быть другая причина. Логично заполнить эти данные мы не можем, а удалить строки с пропусками означает расстаться с львиной долей данных, что помешает нашему дальнейшему анализу.

Проверим уникальные значения столбца user_score.

```python
data['user_score'].unique()
```

Обнаружилась аббревиатура 'tbd', которая расшифровывается как To Be Determined, то есть "Будет определено". Соответственно оценка пользователей с этой аббревиатурой в столбце означает, что рейтинг еще не определен. Заменим 'tbd' на 'NaN'.

### Изучим столбец rating.

```python
data['rating'].unique()
```

В соответсвие с рейтингом ESRB (ознакомиться с ним можно по ссылке https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board), рейтинг E-"для всех"-это окончательная версия названия, ранее название звучало как "К-А". В связи с этим мы решили объединить эту категорию под общим названием "Е".

Рейтинг RP -"Rating Pending" или "Рейтинг ожидается" означет, что продукт ожидает присвоения рейтинга, поэтому посчитаем возможным объединить рейтинг RP и пропуски в одну категорию "unknown".

```python
data['rating'] = data['rating'].fillna('unknown') # Заполним пропуски. 
```

```python
data['rating'] = data['rating'].replace('RP', 'unknown') # Переименуем RP в unknown.
```

```python
data['rating'] = data['rating'].replace('K-A', 'E') # Переименум К-А в Е.
```

```python
data['rating'].unique() # Проверим, что у нас получилось.
```

Значительное количество пропусков в столбцах critic_score, rating и user_score мы не можем заменить логичными значениями или их восстановить, удалить строки-значит потерять большое количество данных, поэтому решено оставить пропуски.

### Проверим гипотезы.

#### Гипотеза о равенстве средних пользовательских рейтингов платформ Xbox One и РС.

Нулевая гипотеза: средний пользовательский рейтинг платформы Xbox One равен среднему пользовательскому рейтингу PC.

Альтернативная гипотеза: средний пользовательский рейтинг платформы Xbox One не равен среднему пользовательскому рейтингу PC.   

```python
Xbox_array = actual_period_data[(actual_period_data['platform'] == 'XOne') & (actual_period_data['user_score'] > 0)]['user_score']
Xbox_array.mean()
```

```python
PC_array = actual_period_data[(actual_period_data['platform'] == 'PC') & (actual_period_data['user_score'] > 0)]['user_score']
PC_array.mean()
```

```python
Xbox_array.dropna().shape
```

```python
PC_array.dropna().shape
```

Средние выборок немного отличаются, для дополнительной проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам мы будем использовать метод ttest_ind.

```python
alpha = .05

results = st.ttest_ind(
Xbox_array.dropna(),
PC_array.dropna(),
)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Посчитав средние значения наших выборок, мы обнаружили, что они немного отличаются, но проведя тест, мы не смогли отвергнуть нулевую гипотезу об их равенстве, это значит, что различия посчитанных средних были скорее всего случайны. 

#### Гипотеза о различие средних пользовательских рейтингов жанров Action и  Sports.

Нулевая гипотеза: средний пользовательский рейтинг жанра Action равен среднему пользовательскому рейтингу жанра Sports.

Альтернативная гипотеза: средний пользовательский рейтинг жанра Action не равен среднему пользовательскому рейтингу жанра Sports.

```python
action_array = actual_period_data[(actual_period_data['genre'] == 'Action') &  (actual_period_data['user_score'] > 0)]['user_score']
action_array.mean()
```

```python
sports_array = actual_period_data[(actual_period_data['genre'] == 'Sports') &  (actual_period_data['user_score'] > 0)]['user_score']
sports_array.mean()
```

```python
action_array.shape
```

```python
sports_array.shape
```

Средние выборок по расчетам действительно не равны, необходимо дополнительно утвердится во мнении с помощью дополнительной проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам. Мы будем использовать метод ttest_ind. Также учтем параметр equal_var, так как размеры выборок значительно различаются.

```python
alpha = .05

results = st.ttest_ind(
action_array,
sports_array,
equal_var=False
)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

У нас получилось отвергнуть нулевую гипотезу о равенстве средних пользовательских рейтингов жанров Action и  Sports. Вероятность получить одинаковые средние ничтожно мала.
