### Предобработка данных.

#### Переименование столбцов.

Приведем названия столбцов к нижнему регистру и добавим нижнее подчеркивание.

```python
purchases.columns = purchases.columns.str.lower()
```

```python
purchases = purchases.rename(columns = {
    'purchaseid':'purchase_id',
    'purchasedate':'purchase_date',
    'customerid':'customer_id',
    'shopid':'shop_id'
    
}
                     )
```

```python
products = products.rename(columns = {'productID':'product_id'})
```

Переименуем категории клиентов по лояльности.

```python
def categorize(loyalty):
    try:
        if loyalty == 1:
            return 'лояльный'
        elif loyalty == 0:
            return 'не участвует'
    except:
        pass
```

```python
purchases['loyalty_program'] = purchases['loyalty_program'].apply(categorize)
```

#### Работа с пропусками.

Мы выявили пропуски значений в столбце customer_id таблицы purchases. Природа этих пропусков возможно связана с тем, что не всегда покупатель регистрировался, покупал разово, оплатив наличными.

```python
purchases.isna().agg(['mean', 'sum']).T.sort_values(by='mean', ascending=False) # посмотрим доли пропусков
```

На долю пропусков приходится около 35 % id покупателей. Посмотрим на распределение пропусков в разрезе лояльных/нелояльных покупателей.

```python
purchases[purchases['loyalty_program'] == 1.0]['customer_id'].isna().agg(['mean', 'sum'])
```

```python
purchases[purchases['loyalty_program'] == 0.0]['customer_id'].isna().agg(['mean', 'sum'])
```

Все пропуска в id клиентов принадлежат нелояльным клиентам. Это вполне логично. Так как чтобы стать лояльным, клиенту небходимо зарегистрироваться и оплатить участие в программе. Так как у каждого чека есть клиент, то мы можем его идентифицировать, назвав наименованием чека. Каждый раз, совершая новую покупку, этот клиент будет идентифицироваться заново по новому номеру чека. 

```python
purchases['customer_id'] = purchases['customer_id'].fillna(purchases['purchase_id'])
```

```python
purchase_date = data.query('shop_id == "Shop 0"').groupby(['date', 'loyalty_program', 'month', 'week', 'weekday']).agg({'revenue':'sum', 'purchase_id':'nunique', 'customer_id':'nunique'}).reset_index()
purchase_date['month'] = purchase_date['month'].replace({1: 'январь', 2: 'февраль', 12: 'декабрь'})
purchase_date['date'] = pd.to_datetime(purchase_date['date'])
purchase_date['date'] = purchase_date['date'].dt.strftime('%Y-%m-%d')
purchase_date
```

## Проверка статистических гипотез.

### Проверка гипотезы о равенстве средних чеков.

Нулевая гипотеза: средний чек участников программы и остальных покупателей равен.

Альтернативная гипотеза: средний чек участников программы и остальных покупателей - отличается .

```python
purchases_1 = data.query('shop_id == "Shop 0"').pivot_table(index=('purchase_id', 'loyalty_program'), values =  'revenue', aggfunc='sum').reset_index()
purchases_1
```

```python
array_1 = purchases_1[purchases_1['loyalty_program'] == 'лояльный']['revenue']
array_1.mean()
```

```python
array_0 = purchases_1[purchases_1['loyalty_program'] == 'не участвует']['revenue']
array_0.mean()
```

```python
array_1.shape
```

```python
array_0.shape
```

Средние выборок по расчетам не равны. Мы будем использовать метод ttest_ind. Также учтем параметр equal_var, так как размеры выборок значительно различаются.

```python
alpha = .05

results = st.ttest_ind(
array_1,
array_0,
equal_var=False
)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

p-значение значительно меньше alpha, это значит, что между выборками имеется статистическая значимость и средние выборок отличаются. Мы отвергаем нулевую гипотезу.

### Проверка гипотезы о том, что среднее количество заказов на клиента  участников программы и  остальных покупателей равны.

Нулевая гипотеза: среднее количество заказов на клиента участников программы и остальных покупателей равно.

Альтернативная гипотеза: среднее количество заказов на клиента участников программы и остальных покупателей отличается.

```python
purchases_2 = data.query('shop_id == "Shop 0"').pivot_table(index=('customer_id', 'loyalty_program'), values =  'purchase_id', aggfunc='nunique').reset_index()
purchases_2
```

```python
array2_1 = purchases_2[purchases_2['loyalty_program'] == 'лояльный']['purchase_id']
array2_1.mean()
```

Средние выборок по расчетам примерно равны, проверим гипотезу о равенстве среднего. Мы будем использовать метод ttest_ind. Также учтем параметр equal_var, так как размеры выборок значительно различаются.
